```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_microdados_inep = pd.read_csv("microdados_inep_ed_basica_2022.csv", sep= ";", encoding= "latin1")
df_microdados_inep
```

```python
df_microdados_inep.info()
```

```python
columns_drop = ["CO_REGIAO", "CO_UF", "CO_MUNICIPIO", "CO_MESORREGIAO", "CO_MICRORREGIAO", "CO_DISTRITO"]
df_microdados_inep1 = df_microdados_inep.drop(columns_drop, axis=1)
df_microdados_inep1
```

```python
df_microdados_inep1.info()
```

```python
df_microdados_inep1.to_csv("bs_inep_analise.csv", index=False)

# Base final INEP utilizada na análise.
```

```python
# Porcetagem do quantitativo por região.

porct_esc_cads_por_reg = df_microdados_inep1["NO_REGIAO"].value_counts(normalize=True)
porct_esc_cads_por_reg.plot(kind="pie", startangle= 90, autopct="%.2f%%", explode= (0.2,0.1,0.1,0.1,0.1));
plt.title(" Percentual de escolas cadastradas por região (INEP 2022)")
plt.xlabel("")
plt.ylabel("")

porct_esc_cads_por_reg

# Região nordeste com maior quantidade de escolas cadastradas.
```

```python
# Quantitativo de escolas cadastradas por estado segundo INEP 2022.

qnt_esc_cads_por_est = df_microdados_inep1["NO_UF"].value_counts()
qnt_esc_cads_por_est.plot(kind="bar", edgecolor='black');
plt.xlabel("")
plt.ylabel("")
plt.title(" Quantidade de escolas cadastradas por estado (INEP 2022)")

qnt_esc_cads_por_est

# São Paulo é o estado com maior quantidade de escolas cadastradas no Brasil.
```

```python
df_pe_inep = df_microdados_inep1[df_microdados_inep1["SG_UF"] =="PE"]
df_pe_inep

# Dataframe separado só de Pernambuco para gerar analisar a questão de quantidade de escolas por municipios.
```

```python
df_mesorregiao = df_pe_inep["NO_MESORREGIAO"].value_counts()
df_mesorregiao.plot(kind="barh", edgecolor='black');
plt.xlabel("")
plt.ylabel("")
plt.title("Escolas por mesorregião")

df_mesorregiao

# Região metropolitana e agreste pernambucano com o dobro da quantidade de escolas cadastradas.
```
